## Representación gráfica de datos.

Si a la función de trazado se le da un matriz de datos, la usará como coordenadas ene eje vertical, y utilizara el indice de cad apunto de datos en el array como la coordenada horizontal.

```python
plt.plot([1,2,5,7,8,3,1])
plt.show()
```

Tambien se pueden proporcionar 2 matrices una para el eje horizontal, y otra para el eje vertical.

```python
plt.plot([-3,-1,0,4,7],[1,4,6,7,8])
plt.show()
```

Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3,-1,0,4,7],[1,4,6,7,8])
plt.axis([-4,8,0,10])# [Xmin, xmax, ymin, ymax]
plt.show()
```

Se sigue el mismo procedimiento para pintar una función matematica

```python
import numpy as np
x= np.linspace(-2,2,500)
y= x**2

plt.plot(x,y)
plt.show()
```

También puede modificarse el estilo de la gráfica para qu econtenga más información

```python
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y =**2")
plt.grid(True)
plt.show()
```

Pueden superponerse gráficas y cambiar el estilo de las funciones.

```python
import numpy as np
x= np.linspace(-2,2,500)
y= x**2
y2= x+1
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y =**2")
plt.grid(True)

plt.plot(x,y,'b--',x, y2, 'g')
plt.show()
```

```python
# Separando en diferentes lineas las funciones 
import numpy as np
x= np.linspace(-2,2,500)
y= x**2
y2= x+1
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y =**2")
plt.grid(True)

plt.plot(x,y,'b--')
plt.plot(x, y2, 'g')
plt.show()
```

```python
import numpy as np
x= np.linspace(-2,2,500)
y= x**2
y2= x+1
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y =**2")
plt.grid(True)

plt.plot(x,y,'b--', label = "x**2")
plt.plot(x, y2, 'g', label = "x + 1")
plt.legend(loc = "best") #La situa en la mejor localizacion
plt.show()
```

También pueden crearse 2 gráficas que no se superpongan, estas gráficas se organizan en un grid y se denominan subplots.

```python
import numpy as np 
x= np.linspace(-2,2,500)
y = x++2
y2 = x+1

plt.subplot(1,2,1) #1rows, 2 columns, 1st Subplots
plt.plot(x,y,'b--')

plt.subplot(1,2,2)# 1rows,2 columns, 2nd Subplots
plt.plot(x, y2, 'g')

plt.show()
```

```python
#Para que las gráficas nomqueden tan ajustadas, se puede hacer la figura más grande 
plt.figure(figsize = (14,6))
plt.subplot(1,2,1) #1rows, 2 columns, 1st Subplots
plt.plot(x,y,'b--')
plt.title("Square Function",fontsize=14)
plt.xlabel("Chido", fontsize=14)
plt.grid



plt.subplot(1,2,2)# 1rows,2 columns, 2nd Subplots
plt.plot(x, y2, 'g')
plt.title("Lineal Function",fontsize=14)
plt.xlabel("Chido2", fontsize=14)
plt.grid
plt.show()
```

## Histogramas 

```python
data = [1,1.1,1.8,2,2.1,3.2,3,3,3,3]
plt.subplot(211)
plt.hist(data, bins =10, rwidth=0.8)

plt.subplot(212)
plt.hist(data, bins =[1,1.5,2,2.5,3], rwidth = 0.95)
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```
